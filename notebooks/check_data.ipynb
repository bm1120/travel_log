```python
from sqlalchemy import create_engine
import pandas as pd

def read_tab(sql):
    # Define connection parameters using SQLAlchemy
    engine = create_engine("postgresql+psycopg2://myuser:mypassword@postgres_db/mydatabase")

    # Query data from one of the tables
    df = pd.read_sql(sql, engine)

    # Transform column names to uppercase
    df.columns = [x.upper() for x in df.columns]
    
    # Close the connection
    engine.dispose()

    return df
```

```python
sql = """
WITH TRP_CONSUME AS (
SELECT 
    *
FROM 
    TN_MVMN_CONSUME_HIS
WHERE
    PAYMENT_SE='교통비'
)
SELECT 
    TRIP_ID
    ,A.TRAVEL_ID
    ,MVMN_CD_1
    ,MVMN_CD_2
    ,B.PAYMENT_AMT_WON AS PAY_1
    ,C.PAYMENT_AMT_WON AS PAY_1
FROM 
    TN_MOVE_HIS A
    LEFT JOIN TRP_CONSUME B
        ON A.TRAVEL_ID = B.TRAVEL_ID
        AND A.MVMN_CD_1 = B.MVMN_SE
    LEFT JOIN TRP_CONSUME C
        ON A.TRAVEL_ID = C.TRAVEL_ID
        AND A.MVMN_CD_1 = C.MVMN_SE
"""
read_tab(sql)
```
